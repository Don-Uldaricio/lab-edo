```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def region(lst):    
    X,Y=np.meshgrid(np.linspace(lst[0],lst[1],lst[2]), np.linspace(lst[0],lst[1],lst[2]))
    return (X,Y)

def campoDir(f,lst):
    U = 1.0
    V = f

    N = np.sqrt(U ** 2 + V ** 2)#Normalizamos los vectores
    U = U / N
    V = V / N

    plt.quiver(X, Y, U, V) #graficamos los vectores con origen en (x,y) y término (u,v)
    plt.xlim([lst[0], lst[1]]) #limites en eje x
    plt.ylim([lst[0], lst[1]]) #limites en eje y

    plt.xlabel('x') #etiqueta eje x
    plt.ylabel('y') #etiqueta eje y
    plt.grid()# para establecer red en la región
```

$$ y' = e^{-x^2} $$

* c) Justifique si la solución que pasa por $(3,0)$ es una función acotada.

    * Recuerde que: Una función $f:I \subset \mathbb{R} \to \mathbb{R}$ se dice acotada, si existe $M > 0$ tal que
$ |f(x)|< M , \quad \forall x \in I.  $

* d) A partir del campo de direcciones, justifique si existe alguna solución constante.

## Solución:

```python
lst = [-5,5,11]
X,Y = region(lst)
campoDir(np.exp(-X**2),lst)
```
